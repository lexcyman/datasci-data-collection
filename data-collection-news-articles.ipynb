## Data Collection - News Articles
This notebook collection data from March 11- 12, 2021 from [NewsAPI](https://newsapi.org/v2/everything?q=bitcoin&from=2021-03-11&to=2021-03-12&apiKey=0249b5764d13418682ff5f5b48426d43) using `requests` and saving the data in a json file.

#### Import requests library to get any website's HTML code to extract its data from.

```python
import requests
URL = "https://newsapi.org/v2/everything?q=bitcoin&from=2021-03-11&to=2021-03-12&apiKey=0249b5764d13418682ff5f5b48426d43"
```

#### Create a query to pass as parameters to the page request. 

```python
query = { "sortBy": "top", "from": "2021-03-11", "to": "2021-03-12", "apiKey": "0249b5764d13418682ff5f5b48426d43" }
```

#### Load the page to the notebook.

```python
page = requests.get(URL, params = query)
```

#### Transform loaded data into json format. 

```python
article_list = page.json()
article_list
```

#### Get only the articles from the website.

```python
articles = article_list["articles"]
articles
```

#### Save data into a json file. 

```python
import json

filename = "news_articles.json"
news_file = open(filename, 'w')
json.dump(articles, news_file, indent = 2)
news_file.close()
```
